```python
from utils import classification
from tensorflow.keras import optimizers
```

- Start with simple problem of mnist classification for ASL alphabets [dataset](https://www.kaggle.com/datamunge/sign-language-mnist)
- See how it goes, may consider using bigger datasets at higher resolutions
- Save data to disk for datageneration later

```python
train_csv = '/home/nic/data/mnist_ssl/sign_mnist_train.csv'
test_csv = '/home/nic/data/mnist_ssl/sign_mnist_test.csv'
model = classification.Classifier()
model.save_mnist(train_csv, test_csv, save=True)
```

```python
batch_size = 128
model.generate_data(batch_size)
```

```python
%%time
model.train(lr=5e-4,
            optimizer=optimizers.RMSprop,
            epochs=20,
            decay_lr=False,
            save_folder='./results/classification',
            notification = True)
```

- Classifier seems to have good edge detectors to pick out the edges of fingers
